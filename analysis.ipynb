```python
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
log = np.loadtxt('remote-signals.txt',delimiter=',')
```

```python
us = log[:,0]
bits = log[:,1]
```

```python
us
```

```python
plt.step(us, bits, where='mid')
```

```python
plt.figure(figsize=(20,5))
plt.step(us, bits, where='mid')
plt.ylim(-0.1,1.1)
plt.xlim(us[0],us[-1])
plt.show()
```

```python
log2 = np.loadtxt('remote-signals.txt',delimiter=',')[:60]
```

```python
us2 = log2[:,0]
bits2 = log2[:,1]
```

```python
plt.figure(figsize=(20,5))
plt.step(us2, bits2, where='mid')
plt.ylim(-0.1,1.1)
plt.xlim(us2[0],us2[-1])
plt.plot(us, bits, '.')
plt.show()
```
